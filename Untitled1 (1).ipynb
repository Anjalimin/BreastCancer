```python
#droping feature
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
# independent variables
x = df.drop('diagnosis',axis=1)
#dependent variables
y = df.diagnosis
```

```python
from sklearn.preprocessing import LabelEncoder
#creating the object
lb = LabelEncoder()
y = lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42, stratify=y)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
def evaluate(train_y, train_pred, val_y, val_pred):
    train_acc = round(accuracy_score([np.argmax(d) for d in train_y], [np.argmax(d) for d in train_pred]) * 100, 2)
    val_acc = round(accuracy_score([np.argmax(d) for d in val_y], [np.argmax(d) for d in val_pred]) * 100, 2)
    bias = round(100 - train_acc, 2)
    variance = round(train_acc - val_acc, 2)
    print("Training accuracy:", str(train_acc) + "%")
    print("Testing accuracy:", str(val_acc) + "%")
```

```python
ann =Sequential()

#ADDING FIRST HIDDEN LAYER WITH 30 NEURONS, THE INPUT LAYER WILL BE ADDED AUTOMATICALLY,
ann.add(Dense(units = 128,activation = 'relu'))
ann.add(BatchNormalization())
ann.add(Dropout(0.3))

#ADDING 2ND HIDDEN LAYER WITH 30 NEURONS
ann.add(Dense(units = 64,activation = 'relu'))
ann.add(BatchNormalization())
ann.add(Dropout(0.3))

ann.add(Dense(units = 2,activation = 'softmax'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model=ann.fit(x_train, y_train,batch_size = 64, validation_data=(x_test, y_test),epochs =100)
```

```python
y_pred_train = ann.predict(x_train)
y_pred_test = ann.predict(x_test)
```

```python
evaluate(y_train, y_pred_train, y_test, y_pred_test)
```
